```python
# Import pandas and numpy
import pandas as pd
import numpy as np

# Create the DataFrame to plot
df = pd.DataFrame({
   'planets': ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 
   'Saturn', 'Uranus', 'Neptune', 'Pluto'],
   'distance': [0.4, 0.7, 1, 1.5, 5.2, 9.6, 19.2, 30, 39.5]
   }, index=[1, 2, 3, 4, 5, 6, 7, 8, 9])

# Create the plot

df.plot.line(); # uncomment to plot
```

```python
# Generate 100 integers (randomly) between 0 and 100
df = pd.DataFrame(np.random.randint(0,100,size=(100, 1)))


# Plot the histogram with 10 bins
df.plot.hist(bins=10); # uncomment to plot
```

```python
# Generate random integers
df = pd.DataFrame(np.random.rand(50, 2), columns=['a', 'b'])

# Create a scatter plot
df.plot.scatter(x='a', y='b'); #uncomment to plot
```

```python
# Create a pandas Series
numbers = pd.Series([1, 2, 2.5, 3, 3.5, 4, 5])

# Plot the KDE
numbers.plot.kde(); # uncomment to plot
```

```python
s = pd.Series([1, 2, 2.5, 3, 3.5, 4, 5])
ax = s.plot.kde()
```

```python
ax = s.plot.kde(bw_method=0.3)
```

```python
ax = s.plot.kde(bw_method=3)
```

```python
ax = s.plot.kde(ind=[1, 2, 3, 4, 5])
```
